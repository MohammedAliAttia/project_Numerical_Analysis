```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Function to fit a line to the data
def fit_line(x, y):
    coefficients = np.polyfit(x, y, 1)
    return np.poly1d(coefficients)

# Define the linear model function
def linear_model(x, m, b):
    return m * x + b

# Function to fit a curve to the data
def fit_curve(x, y, model_func, initial_params=None):
    # Use curve_fit to find the optimal parameters for the model
    try:
        params, covariance = curve_fit(model_func, x, y, p0=initial_params, maxfev=2000)
        return params
    except RuntimeError:
        print("Optimal parameters not found: Number of calls to function has reached maxfev")
        return None

# Input data from the user
num_points = int(input("Enter the number of data points: "))
x_data = np.zeros(num_points)
y_data = np.zeros(num_points)

print("Enter the x and y values for each data point:")
for i in range(num_points):
    while True:
        try:
            x_data[i] = float(input(f"Enter x value for data point {i + 1}: "))
            y_data[i] = float(input(f"Enter y value for data point {i + 1}: "))
            break  # Exit the loop if the input is valid
        except ValueError:
            print("Invalid input. Please enter a valid numeric value.")

# Fit a linear model to the data
linear_params = fit_curve(x_data, y_data, linear_model)

if linear_params is not None:
    print("The fitted curve is linear (linear regression).")

    # Generate x values for smooth curve
    x_fit = np.linspace(min(x_data), max(x_data), 200)

    # Plot the original data points
    plt.scatter(x_data, y_data, label='Data Points')

    # Plot the fitted linear model
    plt.plot(x_fit, linear_model(x_fit, *linear_params), label='Fitted Linear Model')

    # Add labels and legend
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()

    # Show the plot
    plt.show()
```
